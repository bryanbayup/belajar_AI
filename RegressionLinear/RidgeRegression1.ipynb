```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge
```

```python
# Data
X = np.array([[1], [2], [3], [4], [5]])
Y = np.array([60, 70, 80, 85, 90])
```

```python
# Plot data
plt.scatter(X, Y, color='blue')
plt.xlabel('Jumlah Jam Belajar')
plt.ylabel('Nilai Ujian')
plt.title('Hubungan Jumlah Jam Belajar dan Nilai Ujian')
plt.grid(True)
plt.show()
```

```python
# Ridge Regression
ridge_reg = Ridge(alpha=1.0)  # Alpha disini adalah parameter kompleksitas
ridge_reg.fit(X, Y)
```

```python
# Plot hasil Ridge Regression
plt.scatter(X, Y, color='blue')
plt.plot(X, ridge_reg.predict(X), color='red', linewidth=3)
plt.xlabel('Jumlah Jam Belajar')
plt.ylabel('Nilai Ujian')
plt.title('Ridge Regression')
plt.grid(True)
plt.show()
```
