```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Data
X = np.array([[1], [2], [3], [4], [5]])
Y = np.array([60, 70, 80, 85, 90])
```

```python
# Plot data
plt.scatter(X, Y, color='blue')
plt.xlabel('Jumlah Jam Belajar')
plt.ylabel('Nilai Ujian')
plt.title('Hubungan Jumlah Jam Belajar dan Nilai Ujian')
plt.grid(True)
plt.show()
```

```python
# Regresi Linear dengan OLS
reg = LinearRegression()
reg.fit(X, Y)
```

```python
# Plot garis regresi
plt.scatter(X, Y, color='blue')
plt.plot(X, reg.predict(X), color='red', linewidth=3)
plt.xlabel('Jumlah Jam Belajar')
plt.ylabel('Nilai Ujian')
plt.title('Regresi Linear dengan OLS')
plt.grid(True)
plt.show()
```

```python
# Menampilkan koefisien
print("Intercept (beta_0):", reg.intercept_)
print("Koefisien Kemiringan (beta_1):", reg.coef_[0])
```
