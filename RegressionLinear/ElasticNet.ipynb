```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import ElasticNet
from sklearn.datasets import make_regression
```

```python
# Generate synthetic data
X, y = make_regression(n_samples=100, n_features=2, noise=0.1, random_state=42)
```

```python
# ElasticNet Regression
elastic_net = ElasticNet(alpha=0.1, l1_ratio=0.5)  # alpha=regularization strength, l1_ratio=balance between L1 and L2
```

```python
# Fit model
elastic_net.fit(X, y)
```

```python
# Plot data
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot titik data
ax.scatter(X[:, 0], X[:, 1], y, color='blue')

# Plot garis prediksi
# Buat grid untuk plotting surface
x1_grid, x2_grid = np.meshgrid(np.linspace(min(X[:, 0]), max(X[:, 0]), 100),
                               np.linspace(min(X[:, 1]), max(X[:, 1]), 100))
Z = elastic_net.predict(np.array([x1_grid.flatten(), x2_grid.flatten()]).T)
Z = Z.reshape(x1_grid.shape)
ax.plot_surface(x1_grid, x2_grid, Z, alpha=0.5, color='purple')

ax.set_xlabel('Luas Tanah')
ax.set_ylabel('Luas Bangunan')
ax.set_zlabel('Harga Rumah')
ax.set_title('ElasticNet Regression')

plt.show()
```
