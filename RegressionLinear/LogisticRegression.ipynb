```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
# Data
X = np.array([[1], [2], [3], [4], [5]])  # Jumlah jam belajar
Y = np.array([0, 0, 1, 1, 1])  # 0: Tidak lulus, 1: Lulus
```

```python
# Plot data
plt.scatter(X, Y, color='blue')
plt.xlabel('Jumlah Jam Belajar')
plt.ylabel('Lulus (1) / Tidak Lulus (0)')
plt.title('Hubungan Jumlah Jam Belajar dan Hasil Ujian (Lulus/Tidak)')
plt.grid(True)
plt.show()
```

```python
# Logistic Regression
log_reg = LogisticRegression()
```

```python
# Fit model
log_reg.fit(X, Y)
```

```python
# Plot hasil Logistic Regression
plt.scatter(X, Y, color='blue')
plt.plot(X, log_reg.predict_proba(X)[:,1], color='red', linewidth=3)
plt.xlabel('Jumlah Jam Belajar')
plt.ylabel('Probabilitas Lulus Ujian')
plt.title('Logistic Regression')
plt.grid(True)
plt.show()
```
